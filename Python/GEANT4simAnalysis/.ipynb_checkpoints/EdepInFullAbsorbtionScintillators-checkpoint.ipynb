```python
# filepath = '/Users/erezcohen/Desktop/Software/GEANT4/myProjects/rdecay2scintillator-build/'
filepath = '/Users/erezcohen/Desktop/data/PETsys/Geant4Simulations/EvaluationKit/FullAbsorbtionScintillators/'
filename_NaI_scint = filepath + 'Scintillator' + '_G4_SODIUM_IODIDE_' + 'particles' + '.csv'
filename_Plastic_scint = filepath + 'Scintillator' + '_G4_PLASTIC_SC_VINYLTOLUENE_' + 'particles' + '.csv'
filename_Stilbene_scint = filepath + 'Scintillator' + '_G4_STILBENE_' + 'particles' + '.csv'
filename_LYSO_scint = filepath + 'Scintillator' + '_LYSO_' + 'particles' + '.csv'
```

```python
particles_PSC = pd.read_csv(filename_Plastic_scint);
particles_NaI = pd.read_csv(filename_NaI_scint);
particles_Stilbene = pd.read_csv(filename_Stilbene_scint);
particles_LYSO = pd.read_csv(filename_LYSO_scint);
```

```python
fig,ax = plt.subplots(figsize=(9,6));
for particles,label in zip([particles_PSC,particles_NaI,particles_LYSO], #,particles_Stilbene
                           ['Plastic Vinyltoluene','Na(I)','LYSO']):#,'Stilbene'
    Edep = (particles['Edep(scint-1)/MeV']
            +particles['Edep(scint-2)/MeV']
            +particles['Edep(source-holder)/MeV']
            +particles['Edep(world)/MeV']);
    h,edges = np.histogram(Edep , bins=np.linspace(0.1,1.3,100));
    plt.step( edges[:-1] , h, label=label);    
set_axes(ax,x_label='energy deposit [MeV]',y_label='events',do_add_legend=True,do_add_grid=True,xlim=(0.1,1.3));
```

```python
fig,ax = plt.subplots(figsize=(9,6));
for particles,label in zip([particles_Plastic_scint,particles_NaI_scint,particles_Stilbene_scint],
                           ['Plastic Vinyltoluene','Na(I)','Stilbene']):
#     particle_types = np.unique(particles['PrtclName'])
#     print 'in ',label,'particle types:' , particle_types

    p_init = particles['|p_init|/MeV'];
    Edep = particles['Edep(scint-1)/MeV'];
    plt.scatter( p_init, Edep , marker='.', label=label);
set_axes(ax,x_label='initial momentum [MeV]',y_label='energy deposit [MeV]',do_add_legend=True,do_add_grid=True);
```
