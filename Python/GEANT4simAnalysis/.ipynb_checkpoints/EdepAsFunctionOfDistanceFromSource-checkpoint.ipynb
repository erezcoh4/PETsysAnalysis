```python
import sys; sys.path.insert(0, '/Users/erezcohen/Desktop/PETsys/Software/PETsysAnalysis/Python'); 
from PETsys_analysis_tools import *;
%config InlineBackend.figure_format = 'retina'
```

```python
filepath = '/Users/erezcohen/Desktop/data/PETsys/Geant4Simulations/EvaluationKit/RealSizeScintillators/'
filename,particles_data,events_data = dict(),dict(),dict()
distances = np.array([1,10]) #0,10,20,
for dz in distances:
    particles_data[dz] = pd.read_csv(filepath + 'ScintSourceDistance%dmm_particles.csv'%dz);
    events_data[dz] = pd.read_csv(filepath + 'ScintSourceDistance%dmm_events.csv'%dz);
```

```python
main_figures_path = '/Users/erezcohen/Desktop/PETsys/write-ups/Geant4Simulations/EvaluationKit/'
```

```python
fig = plt.figure(figsize=(16,6));
for dz in distances[0:]:
    particles = particles_data[dz]
    label = 'LYSO %d mm from source'%dz

    for i in [1,2]:
        Edep = particles['Edep(scint-%d)/MeV'%i]
        ax=fig.add_subplot(1,2,i)
        h,edges = np.histogram( Edep , bins=np.linspace(0.1,1.3,100));
        plt.step( edges[:-1] , h, label=label);    
        set_axes(ax,x_label='energy deposit [MeV]',
                 y_label='events',do_add_legend=True,do_add_grid=True,xlim=(0.1,1.3));
        
plt.tight_layout()
```

## events

```python
doHist2D = False; # else KDE scatter plot
doKDEscatter = !doHist2D;
doSmearEdep = True
EdepSigma = 0.02 # MeV

for dz in [10]:
    events = events_data[dz]
    Nevents = np.max(events.eventId)
    label = 'LYSO on both sides and %d mm from source, %dM simulated events'%(dz,Nevents/1e6)
        
    events = events[(events['EdepTot(scint-1)/MeV']>0.01)&(events['EdepTot(scint-2)/MeV']>0.01)]
    Edep1 = np.array(events['EdepTot(scint-1)/MeV'])
    Edep2 = np.array(events['EdepTot(scint-2)/MeV'])
    SmearedEdep1 = np.random.normal(loc=Edep1,scale=EdepSigma)
    SmearedEdep2 = np.random.normal(loc=Edep2,scale=EdepSigma)
    x_label = '$E_{dep}^{R}$ [MeV]'
    y_label = '$E_{dep}^{L}$ [MeV]'


    
    if doSmearEdep:
        x,y = SmearedEdep1,SmearedEdep2
        x_label = 'smeared ' + x_label + ' ($\sigma=%.0f$keV)'%(EdepSigma*1000)
        y_label = 'smeared ' + y_label + ' ($\sigma=%.0f$keV)'%(EdepSigma*1000)
    else:
        x,y = Edep1,Edep2
        
    fig = plt.figure(figsize=(8,6));
    
    if doHist2D:
        ax=fig.add_subplot(1,1,1)
        ax.hist2d(x,y,cmap=white_viridis, bins=(np.linspace(0.,0.8,45),np.linspace(0.,0.8,45)));
    else:
        ax=fig.add_subplot(1,1,1,projection='scatter_density')    
        ax.scatter_density(x,y,cmap=white_viridis, dpi=9, downres_factor=2)
    
    set_axes(ax=ax,x_label=x_label,xlim=(0.,0.8),
             y_label=y_label,ylim=(0.,0.8),
             title=label,
             fontsize=17,do_add_grid=True);     
plt.tight_layout()
save_figure(filename = main_figures_path + 'p1_G4sim_10mm_from_source_22Na_EdepCorrelation.pdf')
```
