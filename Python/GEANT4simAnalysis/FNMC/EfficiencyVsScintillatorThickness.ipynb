# estimate neutron detection efficiency and cross-talk probability as a function of scintillator thickness

```python
import sys; sys.path.insert(0, '/Users/erezcohen/Desktop/PETsys/Software/PETsysAnalysis/Python'); 
from PETsys_analysis_tools import *;
main_figures_path = '/Users/erezcohen/Desktop/PETsys/write-ups/Geant4Simulations/FNMC/'
%config InlineBackend.figure_format = 'retina'
```

```python
main_data_path = '/Users/erezcohen/Desktop/data/PETsys/Geant4Simulations/FNMC/ThicknessImpactOnEfficiency/'
EdepThreshold = 200; # keV
```

```python
scintillator_thickness_arr = np.array([1.,5.,30.,40,70.,100.,200.,300]) # mm
primaries_data = dict()
for t in scintillator_thickness_arr:
    filelabel = 'SamplingBoxSide227.0mm_NFacets6_NCellsPerFacet81_SciThickness%.1fmm'%t
    primaries_data[t] = pd.read_csv( main_data_path + filelabel + '_primaries.csv' )
print 'done loading data'
```

```python
P_detection,P_detection_err_up,P_detection_err_dw = [],[],[]
label_detection = '$\\frac{N_{scintillators}>0}{N_{neutrons}}$'

P_NoCrossTalk,P_NoCrossTalk_err_up,P_NoCrossTalk_err_dw = [],[],[]
label_NoCrossTalk = '$\\frac{N_{scintillators}=1}{N_{scintillators}>0}$'

P_CrossTalk,P_CrossTalk_err_up,P_CrossTalk_err_dw = [],[],[]
label_CrossTalk = '$\\frac{N_{scintillators}>1}{N_{scintillators}>0}$'


# efficiency and PnoCrossTalk
for t in scintillator_thickness_arr:
    
    primaries = primaries_data[t]
    detected = primaries[primaries.NScintillatorsFiredPerPrimary>0]
    detected_once = primaries[primaries.NScintillatorsFiredPerPrimary==1]
    detected_more_than_once = primaries[primaries.NScintillatorsFiredPerPrimary>1]
    
    p = float(len(detected))/len(primaries)
    err = p*np.sqrt(1./len(detected) + 1./len(primaries))
    P_detection.append(p)    
    P_detection_err_dw.append( np.min([p, err]))
    P_detection_err_up.append( np.min([1-p, err]))
    
    p = float(len(detected_once))/len(detected)
    err = p*np.sqrt( 1./len(detected_once) + 1./len(detected))
    P_NoCrossTalk.append(p)
    P_NoCrossTalk_err_dw.append( np.min([p, err]))
    P_NoCrossTalk_err_up.append( np.min([1-p, err]))

    
    p = float(len(detected_more_than_once))/len(detected)
    err = p*np.sqrt(1./len(detected_more_than_once) + 1/len(detected))
    P_CrossTalk.append(p)
    P_CrossTalk_err_dw.append( np.min([p, err]))
    P_CrossTalk_err_up.append( np.min([1-p, err]))

# P_CrossTalk,P_CrossTalk_err_up,P_CrossTalk_err_dw = np.array(P_CrossTalk),np.array(P_CrossTalk_err)
# P_NoCrossTalk,P_NoCrossTalk_err = np.array(P_NoCrossTalk),np.array(P_NoCrossTalk_err)
# P_detection,P_detection_err = np.array(P_detection),np.array(P_detection_err)
print 'done.'
```

```python
# plot
fig=plt.figure(figsize=(9,6));

axL=fig.add_subplot(1,1,1)
plt.errorbar(x=scintillator_thickness_arr,y=P_detection,yerr=[P_detection_err_dw,P_detection_err_up],
             marker='o',
             capthick=2,capsize=2);
# cosmetics
set_axes(ax=axL,x_label='scintillator thickness [mm]',
         y_label='Efficiency = ' + label_detection,
         title='Scintillator thickness impact, for proton knockout with $\Delta E>%.1f$ keV'%EdepThreshold,
         fontsize=17,do_add_grid=True,do_add_legend=False);


colorR = 'red'
axR = axL.twinx()  # instantiate a second axes that shares the same x-axis
plt.errorbar(x=scintillator_thickness_arr,y=P_NoCrossTalk,yerr=[P_NoCrossTalk_err_dw,P_NoCrossTalk_err_up],
             marker='o',color=colorR,
             capthick=2,capsize=2);

# cosmetics
set_axes(ax=axR,x_label='scintillator thickness [mm]',ticks_color=colorR,
         y_label='P(0 cross-talk) = ' + label_NoCrossTalk,
         title='Scintillator thickness impact, for proton knockout with $\Delta E>%.1f$ keV'%EdepThreshold,
         fontsize=17,do_add_grid=True,do_add_legend=False);
for tl in axR.yaxis.get_majorticklabels():
    tl.set_fontsize(17)

plt.tight_layout()
save_figure( main_figures_path + 'ScintillatorThicknessImpactForProtonKnockoutWithDeltaEdep200keV.pdf' );
# plt.errorbar(x=scintillator_thickness_arr,y=P_CrossTalk,yerr=[P_CrossTalk_err_dw,P_CrossTalk_err_up],
#              marker='o',
#              capthick=2,capsize=2,label='cross-talk, ' + label_CrossTalk);
```
