```python
data_path = '/Users/erezcohen/Desktop/data/PETsys/external_trigger/'
ExtTrgData = pd.read_csv(data_path+'trigger_rate0.5kHz_amp1250mV_offset1250mV_dutyCycle50_10sec_withNa22_singles.txt',
                          delimiter='\t',names=['time','charge','channel'])
```

```python
# count the number of events between each trigger
gateIdx = 0
Nevents482, Nevents533 = dict(),dict()
Nevents482_arr, Nevents533_arr = [],[]
for inx,row in ExtTrgData.iterrows():#{
    channel = row.channel
    charge = row.charge
    
    if channel==4128 and charge==1:
        # start gate
        gateIdx = gateIdx + 1
        Nevents482[gateIdx] = 0
        Nevents533[gateIdx] = 0

        
        
    if channel==482:
        Nevents482[gateIdx] = Nevents482[gateIdx] + 1
    if channel==533:
        Nevents533[gateIdx] = Nevents533[gateIdx] + 1
        
        
    if channel==4128 and charge==-1:    
        # end gate
        Nevents482_arr.append(Nevents482[gateIdx])
        Nevents533_arr.append(Nevents533[gateIdx])
#} 
Nevents482_arr,Nevents533_arr = np.array(Nevents482_arr),np.array(Nevents533_arr)
print 'done'
```

```python
def integer_hist(data,**args):
    d = np.diff(np.unique(data)).min()
    left_of_first_bin = data.min() - float(d)/2
    right_of_last_bin = data.max() + float(d)/2
    plt.hist(data, np.arange(left_of_first_bin, right_of_last_bin + d, d) , **args)
```

```python
# plot hisotgrams of number of events per gate
bins=np.linspace(0,7,8)
fig=plt.figure(figsize=(12,6)); 

ax = fig.add_subplot(1,2,1)
integer_hist(Nevents482_arr,edgecolor='k',color=(0.9,0.2,0.3,0.1))
set_axes(ax=ax,x_label='number of events per 1 ms gate',y_label='frequency',remove_ticks_y=False,
         fontsize=20,do_add_grid=True,title='1 mm from 0.8$\mu$C $^{22}$Na source');
ax.set_yscale('log')    


ax = fig.add_subplot(1,2,2)
integer_hist(Nevents533_arr,edgecolor='k',color=(0.2,0.3,0.9,0.1))
set_axes(ax=ax,x_label='number of events per 1 ms gate',y_label='',
         xticks=[0,1,2],remove_ticks_y=False, fontsize=20,do_add_grid=True,
         title='20 mm from 0.8$\mu$C $^{22}$Na  source');
ax.set_yscale('log')    

plt.tight_layout()
save_figure(filename = main_figures_path + 'number_of_single_Na22_events_in_ext_gate.pdf')
```
