```python
Na = 6.02e+23 # 1/mol Avogadro number
mC = 12.0107 # Carbon molar mass gr/mol
mH = 1.00794 # Hydrogen molar mass gr/mol
# stilbene is C14H12 by Scintinel 
# [https://www.inradoptics.com/pdfs/datasheets/InradOptics_Datasheet_Stilbene_Final.pdf]
rho_scintillator = 1.15 # gr/cm3
mScintillator = 180.24 # gr/mol
fC = 14. / (14+12)
fH = 12. / (14+12)
```

```python
# atomic density density of each Isotope
rho_C = rho_scintillator * fC * Na / mC
rho_H = rho_scintillator * fH * Na / mH
print 'atomic density:',rho_C,'Carbon/cm3, and',rho_H,'Hydrogen/cm3'
```

```python
# microscopic cross-sections
# neutron-Carbon total cross section is about 1 barn
# [Ericsson, G. Advanced Neutron Spectroscopy in Fusion Research. J Fusion Energ 38, 330â€“355 (2019). https://doi.org/10.1007/s10894-019-00213-9]
Xsec_micro_C = 1 # barn
# neutron-Hydrogen total cross-section is about 0.5 barn
# [database NEA N ENDF/B-VII.1 using JANIS software]
# [https://en.wikipedia.org/wiki/Neutron_cross_section#/media/File:Common_light_element_moderators,_reflectors_and_absorbers.svg]
Xsec_micro_H = 0.5 # barn
```

```python
# scintillator macroscopic cross-section
barn_2_cm2 = 1e-24
Xsec_macro=(rho_C*Xsec_micro_C + rho_H*Xsec_micro_H)*barn_2_cm2
# Hydrogen-only, Unfortunately C is not useful for us, especially at low neutron energies. 
# Below neutron energy of 4.4 MeV (first excited state of C), the most probable 
# fast neutron interaction with C is elastic scattering. The max energy neutron can 
# transfer by this reaction is 28% of its initial energy (compared to all its energy with H). 
# Thus the recoiling C nuclei will have rather low energies. In addition, the high ionization 
# density of recoiling C ion will cause very high quenching of scintillation light. 
# So basically interaction with C will not produce detectable scintillation and the presence 
# of C has the effect of lowering detection efficiency.
Xsec_macro_H=rho_H*Xsec_micro_H*barn_2_cm2
```

```python
thickness = np.linspace(0,10,100) # cm
neutron_attenuation = np.exp( - Xsec_macro_H * thickness )
neutron_efficiency = 1-neutron_attenuation
```

```python
fig,ax=plt.subplots(figsize=(10,6));
plt.plot(thickness,neutron_efficiency,'--',linewidth=3)
# cosmetics
set_axes(ax=ax,x_label='scintillator thickness [cm]',y_label='neutron interaction probability',
             title='neutron attenuation as a function of scintllator thickness',
             fontsize=17,do_add_grid=True);
plt.tight_layout()
```
