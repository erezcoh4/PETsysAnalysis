# The multiplicity distribution
step over the time-stamp array, open a gate from each time-stamp and ask how many events were counted within this gate

```python
# gate = 20.e-6 # 20 ns
gate = 100.e-6 # 100 ns
delay = 10.e-3 # 10 us
t = np.array(events['time[ms]'])
print len(t),'events'
```

```python
mul_ctr_R,mul_ctr_RA = np.zeros(20),np.zeros(20)
for t1_Idx in range(len(t)):
    t1 = t[t1_Idx]
    t_to_end = t[t1_Idx+1:]
    
    mul_R  = len(t_to_end[(t1<=t_to_end)       & (t_to_end<=t1+gate)])
    mul_RA = len(t_to_end[(t1+delay<=t_to_end) & (t_to_end<=t1+delay+gate)])
    mul_ctr_R[mul_R] = mul_ctr_R[mul_R] + 1
    mul_ctr_RA[mul_RA] = mul_ctr_RA[mul_RA] + 1
            
print('done')
print 'multiplicity counter in R gate:',mul_ctr_R[0:5]
```

```python
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(8,4))
# plt.bar(np.array(range(6)),mul_ctr_R[0:6]);
# plt.bar(np.array(range(6)),mul_ctr_RA[0:6]);
nu = np.array(range(4))
rects1 = ax.bar(nu - width/2, mul_ctr_R[0:len(nu)]/np.sum(mul_ctr_R), width, label='$P_{\\nu}^{R}$', edgecolor='k')
rects2 = ax.bar(nu + width/2, mul_ctr_RA[0:len(nu)]/np.sum(mul_ctr_RA), width, label='$P_{\\nu}^{R+A}$', edgecolor='k')

set_axes(ax,'$\\nu$ [multiplicity]','probability',do_add_grid=True,do_add_legend=True,fontsize=14,xticks=(0,1,2,3,4))
ax.set_yscale('log')
plt.tight_layout()
```

## alternative (direct) way to find singles, doubles, triples...

```python
event_t_ms = np.array(events['time[ms]'])
event_dt_ms = np.diff(np.sort(event_t_ms))
gate_dt_ms = 100.e-6 # 100 ns
```

```python
Ndt = len(event_dt_ms)
Nsingles=len(events)
Ndoubles,Ntriples=0,0
for dtIdx in range(Ndt):
    if (event_dt_ms[dtIdx] < gate_dt_ms):
        Ndoubles = Ndoubles + 1
        if (event_dt_ms[dtIdx+1] < gate_dt_ms):
            Ntriples = Ntriples + 1        
singles,doubles,triples = float(Nsingles)/dT,float(Ndoubles)/dT,float(Ntriples)/dT
print('rates for singles,doubles,triples in Hz')
print(singles,doubles,triples)
```

## What does our SVM method predict the singles,doubles and triples to be?
with vanishing cross-talk probability,
$$
\begin{eqnarray} 
		singles & = & S D_{1} P_{d}\\
		doubles & = & S D_{2} P_{d}^{2} \Lambda_{d}  / T \nonumber\\
		triples  & = & S D_{3} P_{d}^{3} \Lambda_{t}  / T \nonumber
	\end{eqnarray}
$$

where 
	$\Lambda_{d} = \frac{e^{-\lambda T}-1+\lambda T}{\lambda}$,
	$\Lambda_{t}=\frac{4e^{-\lambda T}-e^{-2\lambda T}-3+2\lambda T}{\lambda}$,

and 
	$D_{1}=M_{L}(U(D^{sf}_{1}-1)+1)$,
	$D_{2}=\left(U D^{sf}_{2} + m \left(U\left(D^{sf}_{1}-1\right) + 1\right)  D^{if}_{2} \right)$,

For $^{252}$Cf, $M_L = \bar{\alpha} = 1$
so

$$
\begin{eqnarray} 
		singles & = & F M_{L} P_{d} D^{sf}_{1} \bar{\alpha} = F P_{d} D^{sf}_{1} \\
		doubles & = & \frac{FM_{L}^{2}P_{d}^{2}}{2}  \left( D^{sf}_{2} + m D^{sf}_{1} \bar{\alpha} D^{if}_{2} \right) \frac{2\Lambda_{d}}{T}
        = F P_{d}^{2} D^{sf}_{2} \frac{\Lambda_{d}}{T} \nonumber\\
		triples  & = & F P_{d}^{3} D^{sf}_{3} \frac{\Lambda_{t}}{T} \nonumber
	\end{eqnarray}
$$


```python
# Fission rate = 2.4e12 / 3.76 fissions/s/g = 6.4e5 fissions/s/µg
# Specific activity is 532 Ci/g = 532 µCi/µg
# So the fission rate per activity is (6.4e5 fissions/s/µg) / (532 µCi/µg) = 1203 fissions / s / µCi
# Our source activity is 1µC(±50%) so we expect F = 1203 ± 50% fissions / sec
# for Cf252, Ds1 = 3.75, Ds2 = 11.96, Ds3 = 31.81
halflifetime_Cf252 = 2.645 # yr

m_Cf252_1981 = 5e-6; # gram
m_Cf252_2021 = m_Cf252_1981 / np.power(2,(2021-1981)/halflifetime_Cf252); 
print'Cf252 mass in 2021:',m_Cf252_2021*1e9,'ng'


# activity_Cf252_1981 = m_Cf252_1981 * 532; # sepcific activity is 532 Ci/gr
# activity_Cf252_2021 = activity_Cf252_1981 / np.power(2,(2021-1981)/halflifetime_Cf252); 
# print'Cf252 activity in 2021:',activity_Cf252_2021*1e9,'nCi'

# Neutron emission rate by spontaneous fission = 2.4e12 n/s/g
F = 2.4e12 / D_sf_1 * m_Cf252_2021
print'Cf252 fission rate in 2021:',F,'fissions/sec'

ps = 1e-9; ns = 1e-6; us = 1e-3; ms = 1; sec = 1e3;

Pd = 0.148
D_sf_1 = 3.75
D_sf_2 = 11.96
D_sf_3 = 31.81
tau_dieaway_ms = 2.5*ns # 2.5 ns
lambda_dieaway = 1./tau_dieaway;
T_ms = dT*sec;
lT = lambda_dieaway * T_ms


Lambda_d_over_T = (np.exp(-lT)-1+lT)/lT
Lambda_t_over_T = (4*np.exp(-lT) + np.exp(-2*lT) -3 + 2*lT)/(2*lT)


print 'lambda_dieaway:',lambda_dieaway,'ms^-1'
print 'T:',T_ms,'ms'
print '𝛌T:',lT
print '𝚲d/T: %.1f'%Lambda_d_over_T
print '𝚲t/T: %.1f'%Lambda_t_over_T


S_expected = F * Pd * D_sf_1
S_expected_err = S_expected / 2
print 'expected singles rate: %.1f ± %.1f events/sec'%(S_expected,S_expected_err)

D_expected = F * (Pd*Pd) * D_sf_2 * Lambda_d_over_T
D_expected_err = D_expected / 2
print 'expected doubles rate: %.1f ± %.1f events/sec'%(D_expected,D_expected_err)

T_expected = F * (Pd*Pd*Pd) * D_sf_3 * Lambda_t_over_T
T_expected_err = T_expected / 2
print 'expected triples rate: %.1f ± %.1f events/sec'%(T_expected,T_expected_err)
```
