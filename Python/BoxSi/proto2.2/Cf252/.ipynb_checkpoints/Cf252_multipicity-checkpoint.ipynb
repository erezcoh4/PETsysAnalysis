# The multiplicity distribution
step over the time-stamp array, open a gate from each time-stamp and ask how many events were counted within this gate

```python
gate = 20.e-6 # 20 ns
# gate = 100.e-6 # 100 ns
delay = 10.e-3 # 10 us
t = np.array(events['time[ms]'])
print( len(t),'events')
```

```python
mul_ctr_R,mul_ctr_RA = np.zeros(20),np.zeros(20)
for t1_Idx in range(len(t)):
    t1 = t[t1_Idx]
    t_to_end = t[t1_Idx+1:]
    
    mul_R  = len(t_to_end[(t1<=t_to_end)       & (t_to_end<=t1+gate)])
    mul_RA = len(t_to_end[(t1+delay<=t_to_end) & (t_to_end<=t1+delay+gate)])
    mul_ctr_R[mul_R] = mul_ctr_R[mul_R] + 1
    mul_ctr_RA[mul_RA] = mul_ctr_RA[mul_RA] + 1
            
print('done')
print('multiplicity counter in R gate:',mul_ctr_R[0:5])
```

```python
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(8,4))
# plt.bar(np.array(range(6)),mul_ctr_R[0:6]);
# plt.bar(np.array(range(6)),mul_ctr_RA[0:6]);
nu = np.array(range(4))
rects1 = ax.bar(nu - width/2, mul_ctr_R[0:len(nu)]/np.sum(mul_ctr_R), width, label='$P_{\\nu}^{R}$', edgecolor='k')
rects2 = ax.bar(nu + width/2, mul_ctr_RA[0:len(nu)]/np.sum(mul_ctr_RA), width, label='$P_{\\nu}^{R+A}$', edgecolor='k')

set_axes(ax,'$\\nu$ [multiplicity]','probability',do_add_grid=True,do_add_legend=True,fontsize=14,xticks=(0,1,2,3,4))
ax.set_yscale('log')
plt.tight_layout()
```

## alternative (direct) way to find singles, doubles, triples...

```python
event_t_ms = np.array(events['time[ms]'])
event_dt_ms = np.diff(np.sort(event_t_ms))
gate_dt_ms = 20.e-6 # 20 ns
# gate_dt_ms = 100.e-6 # 100 ns
```

```python
Ndt = len(event_dt_ms)
Nsingles=len(events)
Ndoubles,Ntriples=0,0
for dtIdx in range(Ndt):
    if (event_dt_ms[dtIdx] < gate_dt_ms):
        Ndoubles = Ndoubles + 1
        if (( event_dt_ms[dtIdx] + event_dt_ms[dtIdx+1] ) < gate_dt_ms):
            Ntriples = Ntriples + 1        
singles,doubles,triples = float(Nsingles)/dT,float(Ndoubles)/dT,float(Ntriples)/dT
print('rates for singles,doubles,triples in Hz')
print(singles,doubles,triples)
```

## What does our SVM method predict the singles,doubles and triples to be?
with vanishing cross-talk probability,
$$
\begin{eqnarray} 
		singles & = & S D_{1} P_{d}\\
		doubles & = & S D_{2} P_{d}^{2} f \nonumber \\
		triples  & = & S D_{3} P_{d}^{3} f^2 \Lambda_{t}  / T \nonumber
	\end{eqnarray}
$$

where 
	$f = d\Lambda_{d}/dT = 1 - e^{-\lambda T}$,

and 
	$D_{1}=M_{L}(U(D^{sf}_{1}-1)+1)$,
	$D_{2}=\left(U D^{sf}_{2} + m \left(U\left(D^{sf}_{1}-1\right) + 1\right)  D^{if}_{2} \right)$,

For $^{252}$Cf, $M_L = \bar{\alpha} = 1$
so

$$
\begin{eqnarray} 
		singles & = & F M_{L} P_{d} D^{sf}_{1} \bar{\alpha} = F P_{d} D^{sf}_{1} \\
		doubles & = & \frac{FM_{L}^{2}P_{d}^{2}}{2}  \left( D^{sf}_{2} + m D^{sf}_{1} \bar{\alpha} D^{if}_{2} \right) f
        = F P_{d}^{2} D^{sf}_{2} \frac{\Lambda_{d}}{T} \nonumber\\
		triples  & = & F P_{d}^{3} D^{sf}_{3} f^2 \nonumber
	\end{eqnarray}
$$


```python
# average neutron detection efficiency for a 252Cf energy spectrum
# Pd                   = 0.148 
Pd                   = 0.016 
Pd_err               = 0.001
```

```python
measurement_day = date(2021,8,9) # Aug-9, 2021

# Cf252 halflife
halflifetime_Cf252 = 2.645 # yr

# [https://www.sciencedirect.com/topics/medicine-and-dentistry/californium-252]
specific_activity_Bq_per_g_Cf252 = 2.00e13 
activity_t0_252_Ci = 5e-6 # 5 uCi on t0 (July-1,2021)
activity_t0_252_Bqrl = 185e3 # 185 uCi on t0 (July-1,2021)
Cf252fissionRate, _ , _ = Cf252_activity_per_day( measurement_day )
# add 5% uncertainty on the activity
F,F_err = Cf252fissionRate,0.05*Cf252fissionRate
print('252Cf fission rate in measurement time: %.1f%.1f'%(F,F_err),'fissions/sec')
```

```python
'''
Fission rate = 2.4e12 / 3.76 fissions/s/g = 6.4e5 fissions/s/g
Specific activity is 532 Ci/g = 532 Ci/g
So the fission rate per activity is 
(6.4e5 fissions/s/g) / (532 Ci/g) = 1203 fissions / s / Ci

Our source activity is 5C(5%) so we expect

    F = 5 * 1203  5% = 6015  300 fissions / sec

for Cf252

    Ds1 = 3.75, Ds2 = 11.96, Ds3 = 31.81
'''

D_sf_1,D_sf_2,D_sf_3 = 3.75, 11.96, 31.81
lambda_dieaway       = 0.33  # ns^-1
T_ms                 = dT    # sec
f                    = 1 - np.exp(- lambda_dieaway * (dT*1e9) )
f2                   = f*f

print('lambda_dieaway:',lambda_dieaway,'ns^-1')
print('dT:',dT*1e9,'ns')
print('f = 1-exp(-T):',f)


S_expected     = F * Pd * D_sf_1
S_expected_err = np.sqrt(np.square(F_err * Pd * D_sf_1) 
                         + np.square(F * Pd_err * D_sf_1))
print('expected singles rate: %.1f  %.1f events/sec'%(S_expected,S_expected_err))

D_expected = F * (Pd*Pd) * D_sf_2 * f
D_expected_err = np.sqrt(np.square(F_err * (Pd*Pd) * D_sf_2 * f)                          
                         + np.square(F * 2 * Pd * Pd_err * D_sf_2))
print('expected doubles rate: %.1f  %.1f events/sec'%(D_expected,D_expected_err))

T_expected = F * (Pd*Pd*Pd) * D_sf_3 * f*f
T_expected_err = np.sqrt(np.square(F_err * (Pd*Pd*Pd) * D_sf_3 * f*f) 
                         + np.square(F * 3 * Pd * Pd * Pd_err * D_sf_3))
print('expected triples rate: %.1f  %.1f events/sec'%(T_expected,T_expected_err))
```
