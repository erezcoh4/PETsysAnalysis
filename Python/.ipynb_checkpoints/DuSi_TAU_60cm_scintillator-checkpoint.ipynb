```python
def get_all_differences_R_L(t_L,t_R):#{
    # compute time differences between all combinations of t_L and t_R
#     dt = []
#     for i_R in np.arange(len(t_R)): #{
#         for i_L in np.arange(len(t_L)): #{
#             dt.append(t_L[i_L] - t_R[i_R])
#         #}
#     #}
#     return dt    
    mesh = np.array(np.meshgrid(t_R,t_L))
    combinations = mesh.T.reshape(-1, 2)
    dt = np.diff(combinations)
    return dt
#}
```

```python
# LaserCoinc = pd.read_csv('/Users/erezcohen/Desktop/data/PETsys/DuSi/TAU_60cm_scintillator/Laser_attenuation_22dB/with_scintillator_with_laser_600sec_coinc.dat')
# LaserCoinc = pd.read_csv('/Users/erezcohen/Desktop/data/PETsys/DuSi/TAU_60cm_scintillator/Laser_attenuation_20dB/with_scintillator_with_laser_20dB_30sec_coinc.dat')
# LaserCoinc = pd.read_csv('/Users/erezcohen/Desktop/data/PETsys/DuSi/TAU_60cm_scintillator_Oct15_2020/Laser_attenuation_15dB/with_scintillator_with_laser_15dB_30sec_coinc_64_events.dat')
LaserCoinc = pd.read_csv('/Users/erezcohen/Desktop/data/PETsys/DuSi/TAU_60cm_scintillator_Oct22_2020/laser_att_20p0_60sec_coinc.dat',delimiter='\t',
                             names=['mh-n1','mh-j1','time1','e1','id1','mh-n2','mh-j2','time2','e2','id2'])
# LaserCoinc = pd.read_csv('/Users/erezcohen/Desktop/data/PETsys/DuSi/TAU_60cm_scintillator/Laser_attenuation_10dB/with_scintillator_with_laser_10dB_30sec_coinc.dat')
LaserCoinc[0:4]
```

```python
mh_n1,mh_n2 = np.array(LaserCoinc['mh-n1']), np.array(LaserCoinc['mh-n2'])
time1,time2 = np.array(LaserCoinc['time1']), np.array(LaserCoinc['time2'])
e1,e2 = np.array(LaserCoinc['e1']),np.array(LaserCoinc['e2'])
e1
```

```python
fdebug = 0
megaevent = 0

LaserEvents = pd.DataFrame();
Nmax = len(LaserCoinc)
print 'processing',Nmax,'lines of coincidence file'
t_R,t_L,summed_Q = [],[],[]

N_R,N_L = mh_n1[0], mh_n2[0];
N_lines_in_event = N_R + N_L - 1
i_line_in_event = 0;
for i in np.arange(Nmax):         
    
    if fdebug>0: print 'i:',i,',i_line_in_event:',i_line_in_event    
    if i_line_in_event<N_lines_in_event:
        t_R.append(time1[i])
        t_L.append(time2[i])
        summed_Q.append(e1[i]+e2[i])
        if fdebug>1: print 'N_R:',N_R,'N_L:',N_L,'N_lines_in_event:',N_lines_in_event,i_line_in_event,'time1[i]:',time1[i]
        i_line_in_event = i_line_in_event + 1    
        
    if i_line_in_event>=N_lines_in_event:
        # keep only unique values of t_L and t_R
        t_R = np.unique(t_R)
        t_L = np.unique(t_L)
        dt_all = get_all_differences_R_L(t_L,t_R);
        LaserEvents = LaserEvents.append(pd.DataFrame({'Qtotal':np.sum(summed_Q),
                                                       'Qavg':np.sum(summed_Q)/(len(summed_Q)),
                                                         'mean_dt_all':np.mean(dt_all)}, 
                                                        index=[megaevent]), 
                                           ignore_index=True)
        t_R,t_L,summed_Q = [],[],[]                
        N_R,N_L = mh_n1[i],mh_n2[i]
        N_lines_in_event = N_R + N_L - 1
        i_line_in_event = 0;
        
    if i%100000==0:  print 100.*float(i)/Nmax,'%'
print 'done., collected ',len(LaserEvents),' laser events'        
LaserEvents[0:15]
```

```python
fig=plt.figure(figsize=(20,12));
for i,dt,datalabel in zip([1,2,3,4],
                                [LaserEvents.mean_dt_all,LaserEvents.mean_dt_first_10,LaserEvents.mean_dt_first_5,LaserEvents.dt_first],
                                ['average over all combinations','average over combinations of first 10 photons from each side','average over first 5 photons from each side','difference between first photon from each side']):
    ax=fig.add_subplot(2,2,i)
    histo,histbins,patches = plt.hist( dt, bins=np.linspace(-300,1500,100) , fc='royalblue', normed=1, edgecolor='black');
    
    # fit to Gaussian
    (mu, sigma) = norm.fit(dt[abs(dt-np.mean(dt))<300])
    # add a 'best fit' line
    x_fit = np.linspace(-300,1500,1000)
    y_fit = mlab.normpdf( x_fit, mu, sigma)
    l = plt.plot(x_fit, y_fit, 'r--', linewidth=2,label='Gaussian, $\sigma$=%.1f ps'%(sigma))

    # cosmetics
    set_axes(ax=ax,x_label='$t_{left} - t_{right}$ [ps]',y_label='frequency',
             title=datalabel+', $\sigma$=%.1f ps'%sigma,
             fontsize=17,do_add_grid=True,do_add_legend=False,legend_loc='upper left');
    
    # conversion to distance
    c = 0.019 # m/ps for parallel (scattered) transmission [https://www.bnl.gov/phobos/Detectors/Counting_House/Documentation/ToF/general.htm]
    print 'mean L-R for ',datalabel,':',mu*c,'+/',sigma*c,'cm'
plt.tight_layout()
```

```python
fdebug = 0
megaevent = 0

LaserEvents = pd.DataFrame();
Nmax = len(LaserCoinc)/10
print 'processing',Nmax,'lines of coincidence file'
t_R,t_L,summed_Q = [],[],[]

N_R = LaserCoinc['mh-n1'][0];
N_L = LaserCoinc['mh-n2'][0];
i_line_in_event = 0;

for i in np.arange(Nmax): 
    
    N_lines_in_event = N_R + N_L - 1
    i_line_in_event = i_line_in_event + 1
    if fdebug>0: print 'i:',i,',i_line_in_event:',i_line_in_event
    
    if i_line_in_event<N_lines_in_event:
        t_R.append(LaserCoinc['time1'][i])
        t_L.append(LaserCoinc['time2'][i])
        summed_Q.append(float(LaserCoinc['e1'][i])+float(LaserCoinc['e2'][i]))
        
    if i_line_in_event>=N_lines_in_event:
        # keep only unique values of t_L and t_R and sort them by time
        t_R = np.sort(np.unique(t_R))
        t_L = np.sort(np.unique(t_L))
        dt_all = get_all_differences_R_L(t_L,t_R);
        dt_first_10 = get_all_differences_R_L(t_L[0:10],t_R[0:10]);
        dt_first_5 = get_all_differences_R_L(t_L[0:5],t_R[0:5]);
        dt_first_1 = t_L[0] - t_R[0];
        mean_t_R = np.mean(t_R)     
        mean_t_L = np.mean(t_L)     
        dt_mean_L_R = mean_t_L - mean_t_R;

        LaserEvents = LaserEvents.append(pd.DataFrame({'Qtotal':np.sum(summed_Q),
                                                       'Qavg':np.sum(summed_Q)/(len(summed_Q)),
                                                         'mean_dt_L_dt_R':dt_mean_L_R,
                                                         'mean_dt_all':np.mean(dt_all),
                                                         'mean_dt_first_10':np.mean(dt_first_10),
                                                         'mean_dt_first_5':np.mean(dt_first_5),
                                                         'dt_first':dt_first_1
                                                      }, 
                                                        index=[megaevent]), 
                                           ignore_index=True)
        t_R,t_L,summed_Q = [],[],[]        
        i_line_in_event = 0;
        N_R,N_L = LaserCoinc['mh-n1'][i],LaserCoinc['mh-n2'][i]

    if i%10000==0:  print 100.*float(i)/Nmax,'%'
print 'done., collected ',len(LaserEvents),' laser events'        
LaserEvents[0:15]
```
