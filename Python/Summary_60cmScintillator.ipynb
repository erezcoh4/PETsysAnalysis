```python
main_data_path = '/Users/erezcohen/Desktop/data/PETsys/DuSi/summaryFiles/'
filenames = os.listdir(main_data_path)
LaserEventsDict = dict()
for filename in filenames:
    filename_parts = filename.split('_')
    attenuation = float(filename_parts[2][0:-3])
    datename = filename_parts[0]
    LaserEvents = pd.read_csv( main_data_path + filename )
    LaserEventsDict[filename] = LaserEvents
    print datename,attenuation,'dB'
print 'done.'
```

```python
def get_datelabel(datename,Oct15seen=False,Oct22seen=False,Oct29seen=False):
    if datename=='15Oct2020':
        color = 'orange'
        if Oct15seen: datelabel = None
        else: 
            datelabel ='Oct-15'
            Oct15seen = True

    if datename=='22Oct2020':
        color = 'blue'
        if Oct22seen: datelabel = None
        else: 
            datelabel = 'Oct-22'
            Oct22seen = True

    if datename=='29Oct2020':
        color = 'red'
        if Oct29seen: datelabel = None
        else: 
            datelabel = 'Oct-29' 
            Oct29seen = True
            
    return datelabel,color,Oct15seen,Oct22seen,Oct29seen
```

```python
# time resolutions
bins=np.linspace(-5000,3000,100)

sigmas = dict()
sigmas_err = dict()

IdcsToUse = [0,1,4,5,10,15,23,24]
LaserEventKeys = LaserEventsDict.keys()
LaserEventValues = LaserEventsDict.values()

for measIdx in IdcsToUse:
    key = LaserEventsDict.keys()[measIdx]
    LaserEvents = LaserEventsDict.values()[measIdx]
    
    filename_parts = key.split('_')
    attenuation = float(filename_parts[2][0:-3])
    datename = filename_parts[0]      
    datalabel = 'laser @ %.1f dB'%attenuation + ', ' + datename
    dt = LaserEvents.mean_dt_all
    # move to center
    dt = dt[(dt>-1e5) & (dt<1e5)]
    dt = dt-np.mean(dt)
    # omit outliers
    dt = dt[(dt>-1e4) & (dt<1e4)]
    
    if len(dt)>100: 
        dt_2_fit = dt[abs(dt-np.mean(dt))<1e4];
        (mu, sigma) = norm.fit(dt_2_fit)        
        print mu, sigma
        sigmas[key] = sigma
        if len(dt)>1000:  sigma_err = 0.05*sigma
        else: sigma_err = 0.2*sigma
        sigmas_err[key] = sigma_err
        
        # plot and add a 'best fit' line
        fig = plt.figure(figsize=(6,4));
        ax=fig.add_subplot(111)
        
        bins=np.linspace(np.percentile(dt,1)-500,np.percentile(dt,99)+500,100)
        x_fit = np.linspace(np.min(bins),np.max(bins),100)
        y_fit = mlab.normpdf( x_fit, mu, sigma)
        histo,histbins,patches = plt.hist( dt, bins=bins,  normed=1, edgecolor='black');
        plt.plot(x_fit, y_fit, '--r', linewidth=2,label=None)
        
        # cosmetics
        set_axes(ax=ax,x_label='$t_{left} - t_{right}$ [ps]',y_label='',remove_ticks_y=True,
             xlim=(np.min(bins),np.max(bins)),
             title=datalabel + ', $\sigma$=%.1f$\pm$%.1f ps'%(sigma,sigma_err),
             fontsize=17,do_add_grid=True,do_add_legend=False);
        
        plt.tight_layout()
        save_figure(filename = main_figures_path + 'GaussianFit_measurement_%d.pdf'%measIdx)        
    else:
        print 'not enough events in ',datalabel
        
    
print sigmas
print sigmas_err
```
