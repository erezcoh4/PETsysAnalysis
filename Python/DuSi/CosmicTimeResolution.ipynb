```python
def get_all_differences_R_L(t_L,t_R):#{
    # compute time differences between all combinations of t_L and t_R
    mesh = np.array(np.meshgrid(t_R,t_L))
    combinations = mesh.T.reshape(-1, 2)
    dt = np.diff(combinations)
    return dt
#}
```

```python
sec = 1.e12
nanosec = sec*1e-9
microsec = sec*1e-6
milisec = sec*1e-3
event_timewindow = 100*microsec;
```

```python
filepath = '/Users/erezcohen/Desktop/data/PETsys/DuSi/TAU_60cm_scintillator_Nov19_2020/'
# filename = 'cosmic_with_trigger_-300nsdelay_th_e8_100sec_single.dat'
cosmic = pd.concat([pd.read_csv(filepath+'cosmic_th_e8_1800sec_single.dat',delimiter='\t',names=['time','charge','channel']),
                   pd.read_csv(filepath+'cosmic_th_e9_3600sec_single.dat',delimiter='\t',names=['time','charge','channel'])])
# cosmic = pd.read_csv(filepath+'cosmic_th_e9_3600sec_single.dat',delimiter='\t',names=['time','charge','channel'])

cosmic_trigg = cosmic[cosmic.channel==4128]
cosmic_scint = cosmic[cosmic.channel!=4128]
print len(cosmic_scint),'SiPM fires which are not in cosmic trigger'
print len(cosmic_trigg),'cosmic trigger events'
# cosmic
```

```python
# find cosmic events
fdebug = 0
# group events of SiPM fired within a narrow timewindow 
cosmic_events = dict()
scint_t = np.array(cosmic_scint.time); scint_dt = np.concatenate([[0],np.diff(scint_t)])
scint_ch = np.array(cosmic_scint.channel); scint_q = np.array(cosmic_scint.charge);
# cosmic events triggered by fingers 
t_FingerCosmic_arr = cosmic_trigg.time
event_idx,N_SiPMs_registered,NfingerCorrelation = 0,0,0;
event_t_R,event_ch_R,event_q_R,event_t_L,event_ch_L,event_q_L,event_t,event_dt,event_ch,event_q = [],[],[],[], [],[],[],[],[],[]
i = 0
while i < len(scint_t)-1:
    t = scint_t[i]; ch = scint_ch[i]; q = scint_q[i]    # dt = scint_dt[i]; 
    dt = scint_t[i+1] - scint_t[i]
    if (64*6<=ch) and (ch<64*7): side_str = 'R';
    elif (64*8<=ch) and (ch<64*9): side_str = 'L'; # elif (64*9<=ch) and (ch<64*10):
    else: side_str='no side';
    if fdebug>2: print i,': SiPM %d (%s) fired at time %.1f ms, next SiPM (%d) fired after %.1f ns'%(ch,side_str,t/milisec,scint_ch[i+1],dt/nanosec)    
    event_t.append(t); event_dt.append(dt); event_ch.append(ch); event_q.append(q)
    if side_str == 'R':
        event_t_R.append(t); event_ch_R.append(ch); event_q_R.append(q)        
    elif side_str == 'L': 
        event_t_L.append(t); event_ch_L.append(ch); event_q_L.append(q)      
    if np.abs(dt)>event_timewindow:#{
        if fdebug>1: print '---------------------------------------------------------------------'
        N_R = len(event_ch_R); N_L = len(event_ch_L);        
        if fdebug>1: print '%d L and %d R SiPMs collected'%(N_L,N_R)
        if fdebug>1: print event_ch_R, event_ch_L
        # save event if number of SiPMs that fired is greater than 1
        if len(event_ch_L)>0 and len(event_ch_R)>0: #{
            N_SiPMs_registered += N_R + N_L
            dt_all_combinations = get_all_differences_R_L(event_t_L,event_t_R)            
            event = dict({  'ch':event_ch,'q':event_q,'t':event_t,
                            'dt':event_dt,'N_R':len(event_ch_R),'N_L':len(event_ch_L),
                            'ch_R':event_ch_R,'q_R':event_q_R,'t_R':event_t_R,
                            'ch_L':event_ch_L,'q_L':event_q_L,'t_L':event_t_L,
                            'Qtot':np.sum(event_q_L) + np.sum(event_q_R),
                            'Qavg':(np.sum(event_q_L) + np.sum(event_q_R))/(len(event_ch_L)+len(event_ch_R)),
                            'dt_all_combinations':dt_all_combinations,
                            'mean(dt)':np.nanmean(dt_all_combinations),
                            'first t(L) - first t(R)':(np.min(event_t_L) - np.min(event_t_R))
                         })
            event['event time'] = np.min(event_t);
            # check if this event is related to any cosmic event triggered by the fingers
            event['isFingerTrigger'] = 0
            for t_FingerCosmic in t_FingerCosmic_arr:
                if np.abs(t_FingerCosmic - event['event time']) < event_timewindow: 
                    event['isFingerTrigger'] = 1; 
            if event['isFingerTrigger']==1: NfingerCorrelation += 1;
            cosmic_events[event_idx] = event            
            if fdebug: print 'cosmic event',event_idx,'registered at %.2f sec'%(event['event time']/sec),',',N_R,'right SiPMs,',N_L,'left SiPMs'#             print event['ch']#             print event['ch_R'],event['ch_L']
            if fdebug>1: print '---------------------------------------------------------------------'
            event_idx += 1
        #}        
        # clean arrays anyway
        event_t,event_dt,event_ch,event_q = [],[],[],[]
        event_t_R,event_ch_R,event_q_R,event_t_L,event_ch_L,event_q_L = [],[],[],[],[],[]
        event_NSiPMs = 0
    #}             
    i = i+1    # step    
print 'done. collected %d cosmic events from %d SiPMs in total (%.2f of SiPMs that fired), %d correlate with fingers'%(event_idx-1,N_SiPMs_registered,float(N_SiPMs_registered)/len(cosmic_scint),NfingerCorrelation)
```

```python
mean_dt,first_dt,Qtot = [],[],[]
mean_dt_fingers,first_dt_fingers,Qtot_fingers = [],[],[]
for event in cosmic_events.values():
    mean_dt.append(event['mean(dt)'])
    first_dt.append(event['first t(L) - first t(R)'])
    Qtot.append(event['Qtot'])
    if event['isFingerTrigger']==1:
        mean_dt_fingers.append(event['mean(dt)'])
        first_dt_fingers.append(event['first t(L) - first t(R)'])
        Qtot_fingers.append(event['Qtot'])

mean_dt = np.array(mean_dt)
first_dt = np.array(first_dt)
Qtot = np.array(Qtot)
mean_dt_fingers = np.array(mean_dt_fingers)
first_dt_fingers = np.array(first_dt_fingers)
Qtot_fingers = np.array(Qtot_fingers)
```
